```python
from hidroclabc import HidroCLVariable
from hidroclabc import linear_regression_imputation as lri
import matplotlib.ticker as mtick

ndvi = HidroCLVariable('ndvi','/Users/aldotapia/Downloads/veg_o_modis_ndvi_mean_b_d16_p0d.csv')
to_imput = ndvi.observations#[ndvi.observations.columns[range(74,82)]]
```

```python
#to_imput
df = to_imput.drop(to_imput.columns[0], axis=1).notnull().sum().divide(len(to_imput.index)).multiply(100)
ax = df.plot(title = 'Valid observations by catchment', ylim = (0,105), color = 'lightseagreen')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
```
